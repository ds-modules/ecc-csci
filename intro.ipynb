## 5. Edit Mode, Command Mode, and Shortcuts

Jupyter has two modes:

- **Edit mode** — cursor is inside the cell; you are typing
- **Command mode** — cell is selected, but you are not typing inside it

| Action | How |
|--------|-----|
| Enter Edit mode | Press `Enter`, or click inside the cell |
| Enter Command mode | Press `Esc` |

`pandas.ipynb` expects you to use these Command-mode shortcuts:

| Shortcut | Action |
|----------|--------|
| `A` | Insert cell **above** |
| `B` | Insert cell **below** |
| `D`, `D` | Delete selected cell (press D twice) |
| `M` | Change cell to **Markdown** |
| `Y` | Change cell to **Code** |

> Tip: Press `Esc` first, then the shortcut.

## 8. Cell Order Matters

Notebooks remember variables from cells you already ran.

> If a later cell uses a name from an earlier cell, you must run the earlier cell first.

This is critical in `pandas.ipynb`: import cells and `read_csv` cells must run before practice cells that use those DataFrames.

Run these two cells **in order**:

````python
course = "CSCI"
print("Saved the course name.")
````

````python
print("Welcome to", course)
````

If you skip the first cell, the second cell raises `NameError` because `course` was never defined.

When things get confusing, use:

**Kernel → Restart & Run All**

That clears memory and re-runs every cell from the top — exactly what you should do before trusting a finished notebook.

## 9. The Kernel (Very Important for pandas.ipynb)

The **kernel** is the Python engine behind your notebook.

| Action | When to use it |
|--------|----------------|
| **Interrupt** | A cell is stuck |
| **Restart** | Clear all variables |
| **Restart & Run All** | Verify the whole notebook top-to-bottom |
| **Change Kernel** | Imports fail / wrong Python environment |

`pandas.ipynb` imports Pandas and Polars. If you see:

`ModuleNotFoundError: No module named 'polars'`

your notebook is using the wrong Python. Fix it with:

**Kernel → Change Kernel → Python (ecc-csci)**
(or the `.venv` / class environment your instructor named)

Then re-run the import cell.

## 12. Reading Common Errors

You will see errors in `pandas.ipynb`. Read the **last line** first.

| Error | Usual meaning | What to do |
|-------|---------------|------------|
| `NameError` | Variable not defined yet | Run earlier cells, or Restart & Run All |
| `ModuleNotFoundError` | Wrong Python / package missing | Change kernel / install packages |
| `FileNotFoundError` | CSV path / folder is wrong | Check working directory and file names |
| `SyntaxError` | Typo in Python code | Fix quotes, parentheses, commas |

Errors are normal while learning. Fix the cause, then re-run the cell.
